```python
model = xgb.Booster()

model.load_model("/home/ubuntu/joel/workspace/example_project/prod_00/admin@nvidia.com/transfer/cc85e6f1-abb4-461f-80a5-967c9f35e199/workspace/app_server/xgboost_model.json")

X = df.drop(df.columns[0], axis=1)
y = df[df.columns[0]]
```
